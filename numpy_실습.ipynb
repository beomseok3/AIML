```python
#np_array vs list
import numpy as np

# 0~10 2씩 증가하는 수열

series =list(range(0,11,2))
print(series)

numpy_series = np.array(series)
print(numpy_series) # ','가 없다
```

```python
numpy_array = np.arange(0,11,2) # arange(start, stop, step) // range() 와 동일하 method
print(numpy_array)
```

```python
# 0~1 0.2 씩 증가하는 수열
try:
  list1 = list(range(0,1,0.2))
  print(list1)
except Exception as e:
  print("넘파이 매개 변수에는 정수만 가능하다", e)
```

```python
arr2 = np.arange(0,1,0.2)
print(arr2) # arange 매개변수는 실수 전체 가능
```

```python
a = np.linspace(0.0, 1.1, 5)
print(a) # 0~1.1까지 5개로 분할
```

```python
# Numpy array 는 brodcasting 연산이 가능하다. (병렬 처리)

list_plus = list(range(6))
print(list_plus)

list_plus = list_plus + [6,]
print(list_plus)

# list() +연산자는 concatenate기능을 한다.
```

```python
# Numpy array는 병렬처리를 통한 스칼라 계산이 가능하다.

numpy_array = np.arange(6)
print(numpy_array)

print(numpy_array + 5)
print(numpy_array - 5)
print(numpy_array * 5)
print(numpy_array / 5)
```

```python
# 두 넘파이 배열의 열 크기가 맞으면 병렬 연산이 가능하다.
numpy_array2 = np.array([[1,2,3],[4,5,6]])
print(numpy_array2,"\n")
print(numpy_array2 + [10,20,30]) # shape가 (,3)인 넘파이 배열을 (1,3)으로 바꾸고 1인 차원에서 값을 병렬처리
```

```python
# 두 넘파이 배열은 차원은 같으나 배열의 형태가 다르다.
print(np.ndim(numpy_array2), np.ndim(np.array([10,20,30]).reshape(1,3)))
```

```python
# 넘파이 배열 초기화 np.zeros(), np.ones()

zero_array = np.zeros(10) # 1*10 배열 생성, 0.0 으로 초기화한다.
print(zero_array)
```

```python
# 3*4 배열 1 로 초기화

one_arr = np.ones([3,4],dtype=int)
print(one_arr)
```

```python
# 넘파이 배열의 속성 shpae(배열 모양), size(배열 요소 갯수), ndim(차원의 수) dtype( 배열 자료형 )

print(one_arr.shape)
print(one_arr.size)
print(one_arr.ndim)
print(one_arr.dtype)
```

```python
# indexing: idx로 배열 데이터 접근 arr[0] arr[1]
# slicing: 배열의 부분 요소 뽑아내기 arr[start:end]

arr = np.arange(10)
print(arr)

# indexing
print(arr[1])

# slicing
print(arr[3:6]) # start ~ end-1

print(arr[:])

print(arr[3:])

print(arr[:7])
```

```python
# boolean indexing (filter 역할)

bool_index = arr > 5 # 5 초과의 요소에 True를 주는 넘파이 어레이

print(bool_index)

print(arr[bool_index])
```

```python
# 배열의 변형 방법 transpose() 전치행렬
# reshape() 행렬의 shape 바꾸기
# flatten() 1차원 축소

numpy_array3 = np.array([[1,2,3],[4,5,6]])

print(numpy_array3,"\n", numpy_array3.shape)
```

```python
numpy_array3_transpose = numpy_array3.transpose()

print(numpy_array3_transpose)

print(numpy_array3_transpose.shape)
```

```python
f_arr = np.ones([3,4,5], dtype=int)
print(f_arr)
f_arr = np.arange(1, f_arr.size + 1).reshape(f_arr.shape)
print(f_arr.ndim, "\n", f_arr)
```

```python
# 3차원 배열을 1차원으로 축소

f_arr_flatten = f_arr.flatten()
print(len(f_arr_flatten))
print(f_arr_flatten)
```

```python
# reshape(-1,N) 행 크기로 -1이 주어졌다. 배열을 변형할때 열크기를 맞추고 행크기를 정한다.
# reshape 전후로 배열의 size는 동일하게 행 크기를 경정한다.

arr = np.arange(12)
print(arr)

arr_1_12 = arr.reshape(1,12)
print(arr_1_12) # 대괄호가 하나 더 생긴다.

arr_2_6 = arr.reshape(2,6)
print(arr_2_6)

arr_3_4 = arr.reshape(3,4)
print(arr_3_4)

#error
try:
  arr12_45 = arr.reshape(12,45)
except Exception as e:
  print("size가 다르게 변형할 수 없다.",e)

try:
  arr_0_5_24 = arr.reshape(0.5,24)
except Exception as e:
  print("배열의 형태는 정수만 가능하다" , e)
```

```python
# 행의 크기를 고정하지 않는 경우(-1)
arr_negative_1 = arr.reshape(-1,1)
print(arr_negative_1.shape)
print(arr_negative_1)


arr_negative_3 = arr.reshape(-1,3)
print(arr_negative_3.shape)
print(arr_negative_3)

arr_negative_6 = arr.reshape(-1,6)
print(arr_negative_6.shape)
print(arr_negative_6)

arr_negative_12 = arr.reshape(-1,12)
print(arr_negative_12.shape)
print(arr_negative_12)
```
